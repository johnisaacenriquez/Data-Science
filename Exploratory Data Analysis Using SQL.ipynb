## 2.2. SQL Queries

### 2.2.1. Calculate the sum of all salaries

```python
%sql SELECT SUM(SALARY) FROM EMPLOYEES;
```

### 2.2.2. Calculate the average salary

```python
%sql SELECT AVG(SALARY) FROM EMPLOYEES;
```

### 2.2.3. Calculate the average salary rounded to 2 decimal places

```python
%sql SELECT CAST(AVG(SALARY) AS DECIMAL(12,2)) FROM EMPLOYEES;
```

### 2.2.4. Calculate the maximum salary

```python
%sql SELECT MAX(SALARY) FROM EMPLOYEES;
```

### 2.2.5. Display the department names

```python
%sql SELECT DISTINCT(DEP_NAME) FROM DEPARTMENTS;
```

### 2.2.6. Display the job titles in uppercase letters

```python
%sql SELECT UCASE(JOB_TITLE) FROM JOBS;
```

### 2.2.7. Display the length of characters for each job title

```python
%sql SELECT JOB_TITLE, LENGTH(JOB_TITLE) FROM JOBS;
```

### 2.2.8. Display the year when employee E1005 joined the company

```python
%sql SELECT EMPL_ID, YEAR(START_DATE) AS YEAR_JOINED FROM JOB_HISTORY WHERE EMPL_ID = 'E1005';
```

### 2.2.9. Display the Employee ID and Jobs ID of those who joined the company in 2001

```python
%sql SELECT EMPL_ID, JOBS_ID FROM JOB_HISTORY WHERE YEAR(START_DATE) = '2001'
```

### 2.2.10. Display the length of service of employees

```python
%sql SELECT EMPL_ID, (CURRENT DATE - START_DATE) AS SERVICE_LENGTH_YMMDD FROM JOB_HISTORY;
```

### 2.2.11. Display the length of service of employees incuding their last and first names

```python
%sql SELECT JH.EMPL_ID, E.L_NAME, E.F_NAME, (CURRENT DATE - JH.START_DATE) AS SERVICE_LENGTH_YMMDD \
FROM JOB_HISTORY AS JH, EMPLOYEES AS E WHERE JH.EMPL_ID = E.EMP_ID ORDER BY SERVICE_LENGTH_YMMDD DESC;
```

### 2.2.12. List all employees who live in Elgin

```python
%sql SELECT EMP_ID, L_NAME, F_NAME, ADDRESS FROM EMPLOYEES WHERE ADDRESS LIKE '%Elgin%';
```

### 2.2.13. List all employees who were born during 1980's

```python
%sql SELECT EMP_ID, L_NAME, F_NAME, B_DATE FROM EMPLOYEES WHERE YEAR(B_DATE) LIKE '198%' ORDER BY B_DATE;
```

### 2.2.14. List all employees whose salary is between 60000 and 70000

```python
%sql SELECT EMP_ID, L_NAME, F_NAME, SALARY FROM EMPLOYEES WHERE SALARY BETWEEN '60000' AND '70000'
```

### 2.2.15. List all employees ordered by Department ID and within each department ordered by Employee ID

```python
%sql SELECT DEP_ID, EMP_ID, L_NAME, F_NAME FROM EMPLOYEES ORDER BY DEP_ID, EMP_ID;
```

### 2.2.16. List all employees and their department name ordered by Department ID and within each department ordered by Employee ID

```python
%sql SELECT E.DEP_ID, D.DEP_NAME, E.EMP_ID, E.L_NAME, E.F_NAME FROM EMPLOYEES AS E, DEPARTMENTS AS D WHERE E.DEP_ID = D.DEPT_ID_DEP \
ORDER BY E.DEP_ID, E.EMP_ID;
```

### 2.2.17. List all employees, their department name, and their job title ordered by Department ID and within each department ordered by Employee ID

```python
%sql SELECT E.DEP_ID, D.DEP_NAME, E.EMP_ID, E.L_NAME, E.F_NAME, J.JOB_TITLE FROM EMPLOYEES AS E, DEPARTMENTS AS D, JOBS AS J \
WHERE E.DEP_ID = D.DEPT_ID_DEP AND E.JOB_ID = J.JOB_IDENT \
ORDER BY E.DEP_ID, E.EMP_ID;
```

### 2.2.18. List the number of employees in each department and their average salary

```python
%sql SELECT DEP_ID, COUNT(EMP_ID) AS NO_OF_EMPLOYEES, CAST(AVG(SALARY) AS DECIMAL(12,2)) AS AVG_SALARY FROM EMPLOYEES GROUP BY DEP_ID
```

### 2.2.19. List all employees whose salary is lower than the average salary

```python
%sql SELECT EMP_ID, L_NAME, F_NAME, SALARY FROM EMPLOYEES WHERE SALARY < (SELECT AVG(SALARY) FROM EMPLOYEES);
```

### 2.2.20. Execute a column expression that retrieves all employee records with EMP_ID, L_NAME, F_NAME, SALARY, and maximum salary as MAX_SALARY in every row

```python
%sql SELECT EMP_ID, L_NAME, F_NAME, SALARY, \
(SELECT MAX(SALARY) FROM EMPLOYEES) AS MAX_SALARY FROM EMPLOYEES;
```

### 2.2.21. Execute a table expression for the employees tables that excludes columns with sensitive employee data

```python
%sql SELECT * FROM (SELECT EMP_ID, L_NAME, F_NAME, DEP_ID FROM EMPLOYEES);
```

## 3. EDA on SpaceX Launch Data

## 3.1. SpaceX Dataset

```python
%sql SELECT * FROM SPACEXTBL LIMIT 5;
```

## 3.2. SQL Queries

### 3.2.1. Display the unique names of launch sites

```python
%sql SELECT UNIQUE(LAUNCH_SITE) FROM SPACEXTBL;
```

### 3.2.2. Display records where launch site begin with the string "CCA"

```python
%sql SELECT * FROM SPACEXTBL WHERE LAUNCH_SITE LIKE 'CCA%' LIMIT 5
```

### 3.2.3. Display total payload mass carried by boosters launched by NASA (CRS)

```python
%sql SELECT SUM(PAYLOAD_MASS__KG_) AS SUM_PAYLOAD FROM SPACEXTBL WHERE CUSTOMER LIKE 'NASA (CRS)'
```

### 3.2.4. Display average payload mass carried by booster version F9 v1.1

```python
%sql SELECT AVG(PAYLOAD_MASS__KG_) AS AVG_PAYLOAD FROM SPACEXTBL WHERE BOOSTER_VERSION = 'F9 v1.1'
```

### 3.2.5. List the date when the first successfull landing outcome in ground pad was achieved

```python
%sql SELECT MIN(DATE) FROM SPACEXTBL WHERE LANDING__OUTCOME = 'Success (ground pad)'
```

### 3.2.6. List the names of boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000

```python
%sql SELECT BOOSTER_VERSION, PAYLOAD_MASS__KG_ FROM SPACEXTBL WHERE LANDING__OUTCOME = 'Success (drone ship)' AND PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000
```

### 3.2.7. List the total number of successful and failure mission outcomes

```python
%sql SELECT MISSION_OUTCOME, COUNT(MISSION_OUTCOME) AS OUTCOME FROM SPACEXTBL GROUP BY MISSION_OUTCOME
```

### 3.2.8. List the names of the booster versions which have carried the maximum payload mass

```python
%sql SELECT BOOSTER_VERSION, PAYLOAD_MASS__KG_ FROM SPACEXTBL WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL)
```

### 3.2.9. List the failed landing outcomes in drone ship, their booster versions, and launch site names for the year 2015

```python
%sql SELECT DATE, BOOSTER_VERSION, LAUNCH_SITE, LANDING__OUTCOME FROM SPACEXTBL WHERE LANDING__OUTCOME = 'Failure (drone ship)' AND YEAR(DATE) = '2015'
```

### 3.2.10. Rank the count of landing outcomes between the date 2010-06-04 and 2017-03-20 in descending order

```python
%sql SELECT LANDING__OUTCOME, COUNT(*) AS COUNT_LAUNCHES \
FROM SPACEXTBL WHERE DATE BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY LANDING__OUTCOME ORDER BY COUNT_LAUNCHES DESC;
```
